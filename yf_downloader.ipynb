```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
SYMBOLS = ["LIN", "NEM", "GOOG", "META", "MTCH", "TMUS", "NFLX", "AMZN", "EBAY", 
           "TSLA", "GM", "MCD", "WMT", "PG", "KO", "XOM", "COP", "ET", "JPM", "C", 
           "V", "MA", "BX", "BLK", "BRK-B", "LLY", "JNJ", "UNH", "ABT", "VRTX", "GE",
           "RTX", "ETN", "CAT", "PLD", "AMT", "NVDA", "QCOM", "MSFT", "AAPL", "CRM", 
           "UBER", "AVGO", "ORCL", "ADBE", "IBM", "NEE", "SO"]
PERIOD_START = pd.Timestamp(year=2000, month=1, day=1)
PERIOD_END = pd.Timestamp(year=2024, month=7, day=5)
```

```python
lin = yf.download("LIN", start=PERIOD_START, end=PERIOD_END)
lin.to_csv("data/YF_LIN.csv", index=False)
```

```python
def download_sym(sym):
    sym_data = yf.download(sym, start=PERIOD_START, end=PERIOD_END)
    sym_data.to_csv(f"data/YF_{sym}.csv")
```

```python
for sym in SYMBOLS[1:]:
    download_sym(sym)
    print(f"downloaded {sym}")
```
